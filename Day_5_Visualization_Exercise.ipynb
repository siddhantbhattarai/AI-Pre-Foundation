```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/siddhantbhattarai/AI-Pre-Foundation/main/sample_data.csv'
sample_df = pd.read_csv(url)
sample_df.head()
```

```python
sample_df.isnull().sum()
```

```python
clean_df = sample_df.dropna()
```

```python
clean_df.head()
```

```python
# Plot histograms for sample data
clean_df.hist(figsize=(15, 10), bins=10)
plt.suptitle('Histogram for sample data', fontsize=16)
plt.show()
```

```python
# Plot box plot for sample data
plt.figure(figsize=(8,6))
sns.boxplot(data=clean_df.select_dtypes(include=['float64']))
plt.title('Box plots for sample_data')
plt.show()
```

```python
# Plot scatter plots fr sample data
sns.pairplot(clean_df.select_dtypes(include=['float64']))
plt.suptitle('Scatter plots for sample data', y = 1.02)
plt.show()
```

```python
# KDE plot for value in sample data
plt.figure(figsize=(10,7))
sns.kdeplot(clean_df['value'], fill=True, color='blue')
plt.title('KDE plot for value in sample data')
plt.show()
```

```python
# KDE plot for value in sample data
plt.figure(figsize=(10,7))
sns.kdeplot(clean_df['price'], fill=True, color='green')
plt.title('KDE plot for price in sample data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url_1 = 'https://raw.githubusercontent.com/siddhantbhattarai/AI-Pre-Foundation/main/car_sales_data_2500.csv'
new_df = pd.read_csv(url_1)

new_df.head()
```

```python
# 1. Histogram
plt.figure(figsize=(10, 6))
plt.hist(new_df['Sales'], bins=20, edgecolor='black')
plt.title('Distribution of Car Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Manufacturer', y ='Sales', data=new_df)
plt.title('Sales Distribution by Manufacturer')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(new_df['Avg_Price'], new_df['Sales'])
plt.title('Relationship between Average Price and Sales')
plt.xlabel('Average Price')
plt.ylabel('Sales')
plt.show()
```

```python
# 4. Bar plot
yearly_sales = new_df.groupby('Year')['Sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.bar(yearly_sales['Year'], yearly_sales['Sales'])
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 6. Line plot: Yearly trends in average price
yearly_avg_price = new_df.groupby('Year')['Avg_Price'].mean()
plt.figure(figsize=(10, 6))
plt.plot(yearly_avg_price.index, yearly_avg_price.values, marker='o')
plt.title('Yearly Trends in Average Price')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
# 7. Pie chart: Market share based on total sales
plt.figure(figsize=(10, 10))
market_share = new_df.groupby('Manufacturer')['Sales'].sum()
plt.pie(market_share, labels=market_share.index, autopct='%1.1f%%', startangle=90)
plt.title('Market Share Based on Total Sales')
plt.axis('equal')
plt.show()
```

```python
# 8. Bar plot: Average sales by manufacturer
plt.figure(figsize=(12, 6))
avg_sales = new_df.groupby('Manufacturer')['Sales'].mean().sort_values(ascending=False)
avg_sales.plot(kind='bar')
plt.title('Average Sales by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Average Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
