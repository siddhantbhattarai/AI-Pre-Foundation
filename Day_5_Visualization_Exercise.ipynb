```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/siddhantbhattarai/AI-Pre-Foundation/main/sample_data.csv'
sample_df = pd.read_csv(url)
sample_df.head()
```

```python
sample_df.isnull().sum()
```

```python
clean_df = sample_df.dropna()
```

```python
clean_df.head()
```

```python
# Plot histograms for sample data
clean_df.hist(figsize=(15, 10), bins=10)
plt.suptitle('Histogram for sample data', fontsize=16)
plt.show()
```

```python
# Plot box plot for sample data
plt.figure(figsize=(8,6))
sns.boxplot(data=clean_df.select_dtypes(include=['float64']))
plt.title('Box plots for sample_data')
plt.show()
```

```python
# Plot scatter plots fr sample data
sns.pairplot(clean_df.select_dtypes(include=['float64']))
plt.suptitle('Scatter plots for sample data', y = 1.02)
plt.show()
```

```python
# KDE plot for value in sample data
plt.figure(figsize=(10,7))
sns.kdeplot(clean_df['value'], fill=True, color='blue')
plt.title('KDE plot for value in sample data')
plt.show()
```

```python
# KDE plot for value in sample data
plt.figure(figsize=(10,7))
sns.kdeplot(clean_df['price'], fill=True, color='green')
plt.title('KDE plot for price in sample data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url_1 = 'https://raw.githubusercontent.com/siddhantbhattarai/AI-Pre-Foundation/main/car_sales_data_2500.csv'
new_df = pd.read_csv(url_1)

new_df.head()
```

```python
# 1. Histogram
plt.figure(figsize=(10, 6))
plt.hist(new_df['Sales'], bins=20, edgecolor='black')
plt.title('Distribution of Car Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Manufacturer', y ='Sales', data=new_df)
plt.title('Sales Distribution by Manufacturer')
plt.xticks(rotation=45)
plt.show()
```
